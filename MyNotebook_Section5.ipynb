```python
from scipy import spatial
cosin_similarity = lambda vec1,vec2: 1- spatial.distance.cosine(vec1,vec2)
```

```python
king=nlp.vocab['king'].vector
man=nlp.vocab['man'].vector
woman=nlp.vocab['woman'].vector
```

```python
#king-man+woman # New vector similiar to Queen,Princess,Highness
```

```python
new_vector=king-man+woman
```

```python
computed_similarities=[]

for word in nlp.vocab:
    if word.has_vector:
        if word.is_lower:
            if word.is_alpha:
                similarity=cosin_similarity(new_vector,word.vector)
                computed_similarities.append((word,similarity))
```

```python
computed_similarities=sorted(computed_similarities,key=lambda item:-item[1])
```

```python
print([t[0].text for t in computed_similarities[:10]])
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
blanks=[]

for i,label,rev in df.itertuples():
    if type(rev)== str:
        if rev.isspace():
            blanks.append(i)      
len(blanks)
```

```python
df['compound']=df['scores'].apply(lambda d:d['compound'])
```

```python
df['pred_label']=df['compound'].apply(lambda score: 'pos' if score>= 0 else 'neg')
```

```python
df
```
